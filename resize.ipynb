```python
root_dir = '../../data/200003076/'
image_dir = root_dir + 'images/'
label_dir = root_dir + 'labels/' + 'one_x1.0/'
output_image_dir = root_dir + 'images_resized_512/'
output_label_dir = root_dir + 'labels_resized_512/one_x1.0/'
```

```python
if not os.path.exists(output_image_dir):
    os.mkdir(output_image_dir)
```

```python
image_list = os.listdir(image_dir)
```

```python
for image in image_list:
    img = Image.open(image_dir + image)
    img_resize = img.resize((1024,1024),Image.LANCZOS)
    img_resize.save(output_image_dir + image)
```

```python
label_list = os.listdir(label_dir)
if not os.path.exists(output_label_dir):
    os.mkdir(output_label_dir)
print(len(label_list))
```

```python
for label in label_list:
    img = Image.open(label_dir + label)
    img_resize = img.resize((512,512),Image.LANCZOS)
    img_resize.save(output_label_dir + label)
```
